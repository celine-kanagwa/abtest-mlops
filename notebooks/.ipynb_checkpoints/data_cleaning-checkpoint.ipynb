```python
import os
import sys
import pandas as pd
import numpy as np
```

```python
# setting path to folders

parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
scripts_dir = parent_dir+"/scripts"
data_dir = parent_dir+"/data"

sys.path.insert(1, scripts_dir)
```

```python
# loading script

from data_summarizing_functions import DataSummarizer
from dashboard_viz import VizManager

sumar = DataSummarizer()
viz = VizManager()
```

```python
# loading data

original_df = pd.read_csv(data_dir+"/SmartAd_original_data.csv")
```

```python
# lets take a look at missing values

sumar.summ_columns(original_df)
```

```python
# checking if there are rows with 0 input for both yes and no and extracting the opposite.
print ("Missing values: ", original_df.loc[((original_df["yes"]== 0) & (original_df["no"]==0))].shape[0])

clean_df = original_df.loc[~((original_df["yes"]== 0) & (original_df["no"]==0))]

print("Usable rows: ", clean_df.shape[0])
clean_df.head()
```

```python
# saving cleaned data

clean_df.to_csv(data_dir+"/SmartAd_clean_data.csv", index=False)
```
